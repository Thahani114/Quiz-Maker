```python
from tkinter import *
from tkinter import ttk

def next_tab(tab_control):
    tab_control.select(tab2)


def submit_answers(selected1, selected2, selected3, selected4, selected5, selected6):
    correct_answers = {
        1: "7",     
        2: "NewDelhi", 
        3: "Bengaluru", 
        4: "Trivandram", 
        5: "Ernakulam",     
        6: "8"
    }
    user_answers = {
        1: selected1.get(),
        2: selected2.get(),
        3: selected3.get(),
        4: selected4.get(),
        5: selected5.get(),
        6: selected6.get()
    }
    score = 0
    for i in range(1, 7):
        if user_answers[i] == correct_answers[i]:
            score += 1
    results_text = f"Your Score: {score}/6\n\n"
    
    lb.config(text=results_text)
    tab_control.select(tab3)     


window = Tk()
window.title("Quiz")
window.geometry("600x500")

lbl = Label(window, text="Welcome to the Quiz", font=("Helvetica", 12))
lbl.pack(pady=10)
lbl1 = Label(window, text="Uncover your Knowledge", font=("Arial", 10))
lbl1.pack(pady=10)

tab_control = ttk.Notebook(window)
tab_control.pack(expand=1, fill='both')

tab1 = ttk.Frame(tab_control)
tab_control.add(tab1, text='Questions 1-3')

selected1 = StringVar(value="")  
selected2 = StringVar(value="")  
selected3 = StringVar(value="")  

q1 = Label(tab1, text="1. What is 2+5?")
q1.pack(anchor=W)
rad1 = Radiobutton(tab1, text='A) 6', value='6', variable=selected1)
rad2 = Radiobutton(tab1, text='B) 7', value='7', variable=selected1)
rad1.pack(anchor=W)
rad2.pack(anchor=W)

q2 = Label(tab1, text="2. Which among is the capital of India?")
q2.pack(anchor=W)
rad3 = Radiobutton(tab1, text='A) NewDelhi', value='NewDelhi', variable=selected2)
rad4 = Radiobutton(tab1, text='B) Mumbai', value='Mumbai', variable=selected2)
rad3.pack(anchor=W)
rad4.pack(anchor=W)

q3 = Label(tab1, text="3. Which city is known as the Silicon Valley of India?")
q3.pack(anchor=W)
rad5 = Radiobutton(tab1, text='A) Bengaluru', value='Bengaluru', variable=selected3)
rad6 = Radiobutton(tab1, text='B) Hyderabad', value='Hyderabad', variable=selected3)
rad5.pack(anchor=W)
rad6.pack(anchor=W)


Button(tab1, text="Next", command=lambda: next_tab(tab_control)).pack(pady=10)

tab2 = ttk.Frame(tab_control)
tab_control.add(tab2, text='Questions 4-6')

selected4 = StringVar(value="")  
selected5 = StringVar(value="")  
selected6 = StringVar(value="")  

q4 = Label(tab1, text="4. Which among is the capital of Kerala?")
q4.pack(anchor=W)
rad7 = Radiobutton(tab1, text='A) Trivandram', value='Trivandram', variable=selected4)
rad8 = Radiobutton(tab1, text='B) Ernakulam', value='Ernakulam', variable=selected4)
rad7.pack(anchor=W)
rad8.pack(anchor=W)

q5 = Label(tab2, text="5. Which among is known as Commercial capital of Kerala?")
q5.pack(anchor=W)
rad9 = Radiobutton(tab2, text='A) Ernakulam', value='Ernakulam', variable=selected5)
rad10 = Radiobutton(tab2, text='B) Thrissur', value='Thrissur', variable=selected5)
rad9.pack(anchor=W)
rad10.pack(anchor=W)

q6 = Label(tab2, text="6. What is 6+2?")
q6.pack(anchor=W)
rad11 = Radiobutton(tab2, text='A) 8', value='8', variable=selected6)
rad12 = Radiobutton(tab2, text='B) 7', value='7', variable=selected6)
rad11.pack(anchor=W)
rad12.pack(anchor=W)



Button(tab2, text="Submit",command=lambda: submit_answers(selected1, selected2, selected3, selected4, selected5, selected6)).pack(pady=10)
      

tab3 = ttk.Frame(tab_control)
tab_control.add(tab3, text='Results')

lb = Label(tab3, text="", font=("Arial", 12))
lb.pack(pady=20)


window.mainloop()
```
